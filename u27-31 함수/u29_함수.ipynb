## 연숩 문제
- 3자리수 정수 두개를 곱해서 나온 결과가 팔린드롬이 되도록 할 때, 가장 큰 숫자는?

```python
def is_palindrome(x):
    s = str(x)
    return s ==s[::-1] 
```

```python
max_pal = 0
for i in range(100, 1000):
    for k in range(i, 1000):
        if is_palindrome(i*k):
            if i*k > max_pal:
                max_pal = i*k
                m = i
                n =k
print(f'{m} x {n}={max_pal}')
  #튜플 언팩킹 활용
    
```
